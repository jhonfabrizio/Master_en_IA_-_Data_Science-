### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí

# Promedio

esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]

print("Lista de esperanzas de vida: ",esperanzas_de_vida)
lenEsperanzaDeVida = (len(esperanzas_de_vida))
sumaEsperanza = 0


for esperanza in esperanzas_de_vida:
    sumaEsperanza = sumaEsperanza + esperanza

print("Numero de elmentos en la lista de esperanzas de vida: ", round(lenEsperanzaDeVida,2))
print("Suma total de esperanzas de vida; ",round(sumaEsperanza,2))


print("El promedio de esperanza de vida es: ", round(sumaEsperanza/lenEsperanzaDeVida,2))


# Máximo

esperanzaMaxima = 0
for esperanza in esperanzas_de_vida:
    if esperanza > esperanzaMaxima:
        esperanzaMaxima = esperanza

print("La esperanza maxima es: ", round(esperanzaMaxima,2))



# Mínimo

esperanzaMinima = esperanzas_de_vida[1]
for esperanza in esperanzas_de_vida:
    if esperanza < esperanzaMinima:
        esperanzaMinima = esperanza

print("La esperanza minima es: ", round(esperanzaMinima,2))

````

### 2.5 — Agrupaciones

🔧 **Ejercicio 2.5** — Usa `groupby()` para:
1. Calcular la **esperanza de vida promedio** por `region`
2. Calcular el **gasto en salud promedio** por `nivel_ingresos`
3. Calcular el **número de países** por región
4. ¿Qué región tiene la mayor esperanza de vida promedio? ¿Y la menor?

````python
# Tu código aquí
ds = pd.read_csv("/content/sample_data/salud_mundial.csv")
print()
print()
#print("Relleno de valores nulos de columnas numéricas con la media de la columna")
print()
ds = ds.fillna(ds.mean(numeric_only=True))
# 1. Calcular la esperanza de vida promedio por region
promedioEsperanzaVida = ds.groupby("region")["esperanza_vida"].mean()
print("Esperanza de vida promedio por region")
print()
print(promedioEsperanzaVida)
print()
#2. Calcular el gasto en salud promedio por nivel_ingresos
print("Gasto en salud promedio por nivel de ingresos")
print()
promedioGastosalud = ds.groupby("nivel_ingresos")["gasto_salud_usd"].mean()
print(promedioGastosalud)
print()
#3. Calcular el número de países por región
numPaisesRegion = ds.groupby("region")["pais"].count()
print("Numero de paises por region")
print()
print(numPaisesRegion)
print()
# 4. ¿Qué región tiene la mayor esperanza de vida promedio? ¿Y la menor?
RegionMayorEsperanzaVida = ds.groupby("region")["esperanza_vida"].mean().idxmax()
RegionMenorEsperanzaVida = ds.groupby("region")["esperanza_vida"].mean().idxmin()

print("Region con mayor esperanza de vida promedio: ", RegionMayorEsperanzaVida)
print("Region con menor esperanza de vida promedio: ", RegionMenorEsperanzaVida)
print()

````

### 2.6 — Nueva columna calculada

🔧 **Ejercicio 2.6**
1. Crea una nueva columna llamada `categoria_vida` aplicando la función `clasificar_esperanza_vida` que creaste en la Parte 1 a la columna `esperanza_vida`
2. ¿Cuántos países caen en cada categoría? (usa `value_counts()`)
3. ¿Qué porcentaje del total representa cada categoría?

````python
# Tu código aquí
def clasificar_esperanza_vida(años):
  if años < 60:
    return "Baja"
  elif años >= 60 and años <= 74.9:
    return "Media"
  else:
    return "Alta"
  return

ds = pd.read_csv("/content/sample_data/salud_mundial.csv")
print()
print()
#print("Relleno de valores nulos de columnas numéricas con la media de la columna")
print()
ds = ds.fillna(ds.mean(numeric_only=True))

# 1. Crea una nueva columna llamada categoria_vida aplicando la función clasificar_esperanza_vida que creaste en la Parte 1 a la columna esperanza_vida

ds["categoria_vida"] = ds["esperanza_vida"].apply(clasificar_esperanza_vida)
print(ds)

# 2. ¿Cuántos países caen en cada categoría? (usa value_counts())

paisesCantidadCategoriaVida = ds["categoria_vida"].value_counts()
print()
print("Paises por categoria de vida")
print()
print(paisesCantidadCategoriaVida)
print()
porcentajePaisesCantidadCategoriaVida = ds["categoria_vida"].value_counts(normalize=True) * 100
print("Porcentaje de paises por categoria de vida")
print()
print(porcentajePaisesCantidadCategoriaVida)





````
